```python
# Get a Redshift connection from Glue Catalog and retrieving data from Redshift Spectrum
import awswrangler as wr
import boto3
import os
session = boto3.setup_default_session(
    aws_access_key_id='',
    aws_secret_access_key=''
)
os.environ['AWS_DEFAULT_REGION'] = 'us-east-1'
```

```python
# https://aws.amazon.com/premiumsupport/knowledge-center/redshift-cluster-private-public/
con = wr.redshift.connect("rs2")
df = wr.redshift.read_sql_query("SELECT * FROM category", con=con)
con.close()
df
```
